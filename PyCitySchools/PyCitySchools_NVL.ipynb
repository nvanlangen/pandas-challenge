```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
num_schools = school_data["school_name"].nunique()
num_students= student_data["Student ID"].nunique()
total_budget= school_data["budget"].sum()
math_ave = student_data["math_score"].mean()
reading_ave = student_data["reading_score"].mean()
overall_ave = (student_data["reading_score"].mean() +student_data["math_score"].mean())/2
student_data_math = student_data.loc[student_data["math_score"] >= 70,:]
math_pass_pct = len(student_data_math) / len(student_data) * 100
student_data_reading = student_data.loc[student_data["reading_score"] >= 70,:]
reading_pass_pct = len(student_data_reading) / len(student_data) * 100
overall_pct =(reading_pass_pct + math_pass_pct)/2

totals_df = pd.DataFrame({"Number of Schools":[num_schools],
                          "Number of Students":[num_students],
                          "Total Budget":[total_budget],
                          "Math Average":[math_ave],
                          "Reading Average":[reading_ave],
                          "Overall Average":[overall_ave],
                          "Pct. Passing Math":[math_pass_pct],
                          "Pct. Passing Reading":[reading_pass_pct],
                          "Pct. Overall":[overall_pct]})
totals_df
```

```python
school_summary_data = school_data
school_summary_data["BudgetPerStudent"] = school_data["budget"] / school_data["size"]
school_summary_data
```

```python
student_summary_data = student_data.groupby("school_name")
#student_summary_data.sum()
student_summary_mean = student_summary_data.mean()

student_summary_mean = student_summary_mean.drop(columns="Student ID")
student_summary_mean

student_math_pass = student_data.loc[student_data["math_score"] >=70,:]
student_math_pass_temp = student_math_pass.groupby("school_name")
student_math_pass_by_school = student_math_pass_temp.count()
student_math_pass_by_school = student_math_pass_by_school.drop(["Student ID","student_name","gender","grade","reading_score"],axis=1)
student_math_pass_by_school=student_math_pass_by_school.rename(columns={"math_score": "math_passed"})
student_math_pass_by_school
student_reading_pass = student_data.loc[student_data["reading_score"] >=70,:]
student_reading_pass_temp = student_reading_pass.groupby("school_name")
student_reading_pass_by_school = student_reading_pass_temp.count()
student_reading_pass_by_school = student_reading_pass_by_school.drop(["Student ID","student_name","gender","grade","math_score"],axis=1)
student_reading_pass_by_school=student_reading_pass_by_school.rename(columns={"reading_score": "reading_passed"})
student_reading_pass_by_school

student_summary_data_by_school = pd.merge(student_math_pass_by_school,student_reading_pass_by_school,on=["school_name"])
student_summary_data_by_school = pd.merge(student_summary_data_by_school,student_summary_mean,on=["school_name"])
school_summary_data2 = pd.merge(student_summary_data_by_school,school_summary_data,on=["school_name"])
school_summary_data2["math_pct"] = school_summary_data2["math_passed"]/school_summary_data2["size"]*100
school_summary_data2["reading_pct"] = school_summary_data2["reading_passed"]/school_summary_data2["size"]*100
school_summary_data2["overall"] = (school_summary_data2["math_pct"] + school_summary_data2["reading_pct"])/2


school_summary_data2.sort_values("overall",ascending=False)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
grade_9 = student_data.loc[student_data["grade"]=="9th",:]
grade_10 = student_data.loc[student_data["grade"]=="10th",:]
grade_11 = student_data.loc[student_data["grade"]=="11th",:]
grade_12 = student_data.loc[student_data["grade"]=="12th",:]

grade_9_by_school = grade_9.groupby("school_name").mean()

grade_10_by_school = grade_10.groupby("school_name").mean()

grade_11_by_school = grade_11.groupby("school_name").mean()

grade_12_by_school = grade_12.groupby("school_name").mean()

grade_9_by_school = grade_9_by_school.drop(["Student ID","reading_score"],axis=1)
grade_9_by_school = grade_9_by_school.rename(columns={"math_score":"9th Math Score"})

grade_10_by_school = grade_10_by_school.drop(["Student ID","reading_score"],axis=1)
grade_10_by_school = grade_10_by_school.rename(columns={"math_score":"10th Math Score"})

grade_11_by_school = grade_11_by_school.drop(["Student ID","reading_score"],axis=1)
grade_11_by_school = grade_11_by_school.rename(columns={"math_score":"11th Math Score"})

grade_12_by_school = grade_12_by_school.drop(["Student ID","reading_score"],axis=1)
grade_12_by_school = grade_12_by_school.rename(columns={"math_score":"12th Math Score"})

math_by_grade = pd.merge(grade_9_by_school,grade_10_by_school,on=["school_name"])
math_by_grade = math_by_grade.merge(grade_11_by_school,on=["school_name"])
math_by_grade = math_by_grade.merge(grade_12_by_school,on=["school_name"])
math_by_grade
```
